```python
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
NS = {"tei": "http://www.tei-c.org/ns/1.0"}
tree = ET.parse('Corpus/wikis/pt-BR/WIKa.xml')
root = tree.getroot()
```

```python
char_count = 0
full_text = ''
max_count = 1000000 # 1MB

for el in root.findall('.//tei:TEI', NS):
    text_el = el.find('.//tei:text', NS)
    if text_el is None:
        continue

    body = text_el.find('.//tei:body', NS)
    if body is None:
        continue

    for t in body.findall('.//tei:p', NS):
        text = ''.join(t.text).strip()
        char_count += len(text)
        full_text += text

    if char_count >= max_count:
        break

with open("1mb_file.txt", "w") as f:
    f.write(full_text)
```
